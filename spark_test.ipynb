```python
from pyspark.sql import SparkSession

spark = SparkSession \
    .builder \
    .appName("afp_news") \
    .getOrCreate()
```

```python
df = spark.read.json("hdfs:///afp_news_data/news/2019-03-06_AFP_news.json")
df.show()
```

```python
df.printSchema()
```

```python
df.registerTempTable("df");
df.printSchema();
```

```python
from sqlalchemy import create_engine
engine = create_engine("mysql+pymysql://{}:{}@{}/{}?charset={}".format('root', 'root', '35.222.24.169:3306', 'news','utf8mb4'))
con = engine.connect() #创建连接


for i in df:
    app = df.select(i.news_id.alias('news_id'), i.news_link.alias('news_link'), i.img_link.alias('img_link'), i.news_content.alias('news_content'));
    app.registerTempTable("app");
    app_pd = app.toPandas()
    #app_pd 
    app_pd.to_sql(name='afp_news_test', con=con, if_exists='append', index=False)
    #app.show()
```
